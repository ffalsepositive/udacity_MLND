# Model Evaluation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

from sklearn.metrics import classification_report, f1_score
```

## Load Data

### Train

```python
train = pd.read_csv("../data/processed/train.csv", index_col=0)
```

```python
train.head(10).T.head()
```

### Test

```python
test = pd.read_csv("../data/processed/test.csv", index_col=0)
```

```python
test.head(10).T.head()
```

```python
print("Train: ", train.shape)
print("Test: ", test.shape)
```

## Load Models

```python
model_names = ["dummy", "logreg", "knn", "dtree", "rforest", "dtree_hyp", "rforest_hyp"]
bucket = "bucket"
```

```python
models = dict()
for mn in model_names:
    filepath = "../model/" + mn + ".pkl"
    models[mn] = pickle.load(open(filepath, 'rb'))
```

# Performance on Train

```python
feat_cols, target_cols = [i for i in train.columns if i != "target"], ["target"]
X_train, y_train = train[feat_cols].values, train[target_cols].values.ravel()
```

### Dummy Model

```python
dummy = models["dummy"]
y_pred_dummy = dummy.predict(X_train)

print(classification_report(y_train, y_pred_dummy))
```

### Logistic Regression

```python
logreg = models["logreg"]
y_pred_logreg = logreg.predict(X_train)

print(classification_report(y_train, y_pred_logreg))
```

### kNN

```python
knn = models["knn"]
y_pred_knn = knn.predict(X_train)

print(classification_report(y_train, y_pred_knn))
```

### Decision Tree

```python
dtree = models["dtree"]
y_pred_dtree = dtree.predict(X_train)

print(classification_report(y_train, y_pred_dtree))
```

### Random Forest

```python
rforest = models["rforest"]
y_pred_rforest = rforest.predict(X_train)

print(classification_report(y_train, y_pred_rforest))
```

## Hyperparameter Models

### Decision Tree

```python
dtree_hyp = models["dtree_hyp"]
y_pred_dtree_hyp = dtree_hyp.predict(X_train)

print(classification_report(y_train, y_pred_dtree_hyp))
```

### Random Forest

```python
rforest_hyp = models["rforest_hyp"]
y_pred_rforest_hyp = rforest_hyp.predict(X_train)

print(classification_report(y_train, y_pred_rforest_hyp))
```

# Performance on Test

```python
feat_cols, target_cols = [i for i in test.columns if i != "target"], ["target"]
X_test, y_test = test[feat_cols].values, test[target_cols].values.ravel()
```

### Dummy Model

```python
pred_dummy = dummy.predict(X_test)
print(classification_report(y_test, pred_dummy))
```

### Logistic Regression

```python
pred_logreg = logreg.predict(X_test)
print(classification_report(y_test, pred_logreg))
```

#### Feature Importance

```python
feat_imp = pd.concat([pd.DataFrame(feat_cols), \
                      pd.DataFrame(logreg.coef_[0,:]), \
                      pd.DataFrame(logreg.coef_[1,:]), \
                      pd.DataFrame(logreg.coef_[2,:])], axis=1)

feat_imp.columns = ["Feature Name", "Coeff_0", "Coeff_1", "Coeff_2"]
feat_imp.sort_values("Coeff_0", ascending=False)
```

### kNN

```python
pred_knn = knn.predict(X_test)
print(classification_report(y_test, pred_knn))
```

### Decision Tree

```python
pred_dtree = dtree.predict(X_test)
print(classification_report(y_test, pred_dtree))
```

### Random Forest

```python
pred_rforest = rforest.predict(X_test)
print(classification_report(y_test, pred_rforest))
```

## Hyperparameter Models

### Decision Tree

```python
pred_dtree_hyp = dtree_hyp.predict(X_test)
print(classification_report(y_test, pred_dtree_hyp))
```

#### Feature Importance

```python
feat_imp = pd.concat([pd.DataFrame(feat_cols), pd.DataFrame(dtree_hyp.feature_importances_)], axis=1)
feat_imp.columns = ["Feature Name", "Coefficient"]
feat_imp.sort_values("Coefficient", ascending=False)
```

### Random Forest

```python
pred_rforest_hyp = rforest_hyp.predict(X_test)
print(classification_report(y_test, pred_rforest_hyp))
```

#### Feature Importance 

```python
feat_imp = pd.concat([pd.DataFrame(feat_cols), pd.DataFrame(rforest_hyp.feature_importances_)], axis=1)
feat_imp.columns = ["Feature Name", "Coefficient"]
feat_imp.sort_values("Coefficient", ascending=False)
```

# Summary

```python
scores = list()
for mn in model_names:
    preds = eval("y_pred_"+mn)
    s1 = f1_score(y_train, preds, average="weighted")
    s2 = f1_score(y_train, preds, average="macro")
    
    preds = eval("pred_"+mn)
    s3 = f1_score(y_test, preds, average="weighted")
    s4 = f1_score(y_test, preds, average="macro")
    
    scores.append([mn, s1, s2, s3, s4])
    
score_df = pd.DataFrame(scores)
score_df.columns = ["model_name", "train_f1_weighted", "train_f1_macro", "test_f1_weighted", "test_f1_macro"]
score_df
```

```python
# histogram plots for each variable
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 12))
sns.barplot(data=score_df, x="model_name", y="train_f1_weighted", ax=ax1)
ax1.set_title("Weighted F1 Score on Train")

sns.barplot(data=score_df, x="model_name", y="train_f1_macro", ax=ax2)
ax2.set_title("Macro F1 Score on Train")

sns.barplot(data=score_df, x="model_name", y="test_f1_weighted", ax=ax3)
ax3.set_title("Weighted F1 Score on Test")

sns.barplot(data=score_df, x="model_name", y="test_f1_macro", ax=ax4)
ax4.set_title("Macro F1 Score on Test")
```

# Conclusion

- Regarding performance, the fundamental metric is f1_score weighted.
- Weighted f1_score provides the general performance of the model considering sample size of the classes.
- Also macro average is calculated for insight.

### Train Performance
    - Decision Tree & Random Forest without CV performed well on train. It seems that they overfitted.
    - Dummy Classifier is the worst.

### Test Performance
    - The best model is tuned Random Forest Model.
    - Tuned Decision Tree and Random Forest with default parameters are performed well too.
    
### Feature Importance
    - For Tuned Random Forest Model:
        - transaction related features are very important for decisions.
        - reward amount is also important.
        - age and income related features are not very decisive.
    - For Tuned Decision Tree Model:
        - this model cares to similar features as Tuned Random Forest. 
        - also social channel is slightly important.
